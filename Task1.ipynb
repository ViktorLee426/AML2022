## Subtask 0 Filling missing values  
Imputation of missing values  
We can imputate with mean value, median value or via KNN algorithm

```python
#impute null values with median, in X_train and X_test
for i in X_train:
    X_train.loc[X_train.loc[:,i].isnull(),i]=X_train.loc[:,i].median()
for i in X_test:
    X_test.loc[X_test.loc[:,i].isnull(),i]=X_test.loc[:,i].median()
```

```python
#There is no null value anymore
X_train.isna().sum().sum()
```

ABOD: Angle Based Outlier Detection  
This algorithm finds potential outliers by considering the variances of the angles between the data points. It still takes distances into account, but only as a secondary measure to normalize the results.

Feature scaling  Normalization VS Standarization
Normalization is useful when we don’t know about the distribution	  
Standarization is useful when the feature distribution is Normal or Gaussian.

```python
from sklearn import preprocessing
from sklearn.preprocessing import RobustScaler
scaler = preprocessing.RobustScaler().fit(X_train)
scaler
```

```python
X_train_scaled = scaler.transform(X_train)
X_train_scaled
```

```python
X_test_scaled = scaler.transform(X_test)
X_test_scaled 
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns = X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns = X_test.columns)
```

## Subtask 2 Feature Selection   
https://towardsdatascience.com/5-feature-selection-method-from-scikit-learn-you-should-know-ed4d116e4172    
SelectKBest  
The premise with SelectKBest is combining the univariate statistical test with selecting the K-number of features based on the statistical result between the X and y.  


```python
#SelectKBest
from sklearn.feature_selection import SelectKBest, mutual_info_regression
#Select top 80 features based on mutual info regression
selector = SelectKBest(mutual_info_regression, k = 80)
selector.fit(X_train_scaled, y_train)
X_train_scaled.columns[selector.get_support()]
```

```python
selected_features = X_train_scaled.columns[selector.get_support()].tolist()
```

```python
selected_features
```

```python
X_train = X_train_scaled[selected_features]
X_train
```

```python
X_test = X_test_scaled[selected_features]
X_test
```
